```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
df_group = df.groupby('Genre').agg({'NA_Sales':'mean','EU_Sales':'mean','JP_Sales':'mean','Other_Sales':'mean'})

df_group
```

<h2>Soal 2: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
df_trans = df_group.transpose()
#code here

import numpy as np
fig, ax = plt.subplots(figsize=(12,8))

x = np.arange(4)
width = 0.05

warna = ['orangered','dodgerblue','mediumpurple','gray','gold','mediumseagreen','violet','orangered','dodgerblue','mediumpurple','gray','gold']

for i,j in enumerate(df_trans.columns):
    ax.bar(x+(width*i), df_trans[j], width=width, label=j, color=warna[i])


ax.set_title('Means Sales Video Game by Genre')
ax.set_ylabel('Mean Sales')
ax.set_ylabel('Region Sales')
plt.legend()
plt.xticks(x + 0.3 , labels=df_trans.index)
plt.show()


```

Jawab:

Dari grafik saya dapat mengetahui rata-rata penjualan games by genre pada masing masing region kita juga dapat mengetahui mana genre yang paling laku terjual pada masing masing region yg terdapat pada tabel pada tabel ini kita dapat melihat jelas masing-masing rata2 genre game yang terjual dengan jelas pada setiap region

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
fig,ax =plt.subplots(figsize=(12,8))
warna = ['orangered','dodgerblue','mediumpurple','gray','gold','mediumseagreen','violet','orangered','dodgerblue','mediumpurple','gray','gold']
y=[]
for i,j in enumerate(df_trans.columns):
    ax.bar(df_trans.index, df_trans[j], bottom=sum(y),  label=j, color=warna[i])
    y.append(df_trans[j])

ax.set_title('Means Sales Video Game by Genre')
ax.set_ylabel('Mean Sales')
ax.set_ylabel('Region Sales')
plt.legend()
plt.show()
```

Jawab: dari grafik saya bisa melihat total penjualan rata2 dari setiap region disini juga saya bisa tahu mana region yang menjual game paling banyak
